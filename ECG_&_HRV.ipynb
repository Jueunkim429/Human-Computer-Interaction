```python
#initial imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import heartpy as hp
```

```python
#set column names, 첫 번째 colums만 시각화
col_names=['I', 'II', 'III', 'aVR', 'aVL', 'aVF', 'V1', 'V2','V3','V4', 'V5', 'V6']

#load data files
df1 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/실습파일 - Part3-20220920/Sample_SR.csv", names= col_names)
df2 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/실습파일 - Part3-20220920/Sample_SB.csv", names= col_names)
df3 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/실습파일 - Part3-20220920/Sample_ST.csv", names= col_names)
df4 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/실습파일 - Part3-20220920/Sample_AFIB.csv", names= col_names)

sample_rate= 500 #sampling rate
```

```python
# SR
for i in range(12) :
  wd, m =hp.process(df1[col_names[i]], sample_rate)
  hp.plotter(wd,m)
```

```python
filtered = hp.filter_signal(df1[col_names[0]], cutoff = 0.05, sample_rate=sample_rate, filtertype='notch')
```

```python
for i in range(11) :
  wd, m =hp.process(hp.scale_data(filtered), sample_rate)
  hp.plotter(wd,m)
```

```python
# SB
for i in range(11) :
  wd, m =hp.process(df2[col_names[i]], sample_rate)
  plt.figure(figsize=(12,4))
  hp.plotter(wd,m)
```

```python
# ST
for i in range(11) :
  wd, m =hp.process(df3[col_names[i]], sample_rate)
  hp.plotter(wd,m)
```

```python
# AFIB
for i in range(11) :
  wd, m =hp.process(df4[col_names[i]], sample_rate)
  hp.plotter(wd,m)
```
