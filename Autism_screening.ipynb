# 데이터 전처리

```python
data = data.drop(columns=['result','age_desc'], axis=1)
```

```python
x = data.drop(columns='Class/ASD', axis=1)
y=data['Class/ASD']
```

```python
x.loc[:10]
```

 # 원-핫 인코딩

```python
X = pd.get_dummies(x)
X.columns.values
```

```python
X.loc[1]
```

```python
Y=pd.get_dummies(y)
```

# 데이터셋 분리

```python
!pip install sklearn
```

```python
from sklearn import model_selection
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X,Y,test_size=0.2, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

# 신경망 설계

```python
def create_model():
  model = keras.models.Sequential()
  model.add(keras.layers.Dense(8,input_dim=96, kernel_initializer='normal', activation='relu'))
  model.add(keras.layers.Dense(4, kernel_initializer='normal', activation='relu'))
  model.add(keras.layers.Dense(2, activation='softmax'))

  adam = keras.optimizers.Adam(learning_rate=0.001)
  model.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy'])
  return model

model = create_model()
print(model.summary())
```

# 모델 학습

```python
model.fit(X_train, Y_train, epochs=50, batch_size=10, verbose=1)
```

# 신경망 테스트

```python
import numpy as np
from sklearn.metrics import classification_report, accuracy_score

predictions = np.argmax(model.predict(X_test), axis=1)
predictions
```

```python
print('신경망 예측 결과')
print(accuracy_score(Y_test[['YES']], predictions))
print(classification_report(Y_test[['YES']], predictions))
```

# 드롭아웃 규제 추가

```python
dropdout_rate = 0.25

def create_model():
  model = keras.models.Sequential()
  model.add(keras.layers.Dense(8,input_dim=96, kernel_initializer='normal', activation='relu'))
  model.add(keras.layers.Dropout(dropdout_rate))
  model.add(keras.layers.Dense(4, kernel_initializer='normal', activation='relu'))
  model.add(keras.layers.Dropout(dropdout_rate))
  model.add(keras.layers.Dense(2, activation='sigmoid'))
  adam = keras.optimizers.Adam(learning_rate=0.001)

  model.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy'])
  return model
```

# 모델 학습

```python
model = create_model()
model.fit(X_train, Y_train, epochs=50, batch_size=10, verbose=1)
```

# 신경망 테스트

```python
predictions = np.argmax(model.predict(X_test), axis=1)
predictions

print('신경망 예측 결과')
print(accuracy_score(Y_test[['YES']], predictions))
print(classification_report(Y_test[['YES']], predictions))
```
