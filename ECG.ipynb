# 가장 대표적인 4가지 Type에 대한 샘플 데이터를 확인해보기.

```python
#initial imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#set column names, 첫 번째 colums만 시각화
col_names=['I', 'II', 'III', 'aVR', 'aVL', 'aVF', 'V1', 'V2','V3','V4', 'V5', 'V6']

#load data files
df1 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/실습파일 - Part3-20220920/Sample_SR.csv", names= col_names)
df2 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/실습파일 - Part3-20220920/Sample_SB.csv", names= col_names)
df3 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/실습파일 - Part3-20220920/Sample_ST.csv", names= col_names)
df4 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/실습파일 - Part3-20220920/Sample_AFIB.csv", names= col_names)
srate = 500 #sampling rate
timestamps = np.arange(0,10,1/srate) #time in second
x = timestamps #for x axis
```

```python
fig, axs = plt.subplots(4,1,figsize=(10,8))
label_names = ['SR', 'SB', 'ST', 'AFIB']

axs[0].plot(x, df1[col_names[0]])
axs[1].plot(x, df2[col_names[0]])
axs[2].plot(x, df3[col_names[0]])
axs[3].plot(x, df4[col_names[0]])

for i in range(4) :
  axs[i].set_xlabel('Seconds')
  axs[i].set_ylabel('Microvolts')
  axs[i].grid(True)
  axs[i].set_title(label_names[i])
  plt.tight_layout()
```
